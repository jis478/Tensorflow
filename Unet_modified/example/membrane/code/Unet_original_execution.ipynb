```python
from tf_unet import unet, util, image_util

#preparing data loading
data_provider = image_util.ImageDataProvider(search_path = "/notebooks/Unet_membrane/unet/data/membrane/train/new_data/*.png",
                                            a_min=None, a_max=None, data_suffix=".png", mask_suffix='.labels.png', shuffle_data=True)

output_path = '/home/Markkim/temp2/'

#setup & training
net = unet.Unet(cost = 'dice_coefficient', layers=3, features_root=64, channels=1, n_class=2)
trainer = unet.Trainer(net)
path = trainer.train(data_provider, output_path, prediction_path = '/home/Markkim/temp2/', training_iters=32, epochs=40)

# prediction = net.predict(path, data)
# unet.error_rate(prediction, util.crop_to_shape(label, prediction.shape))
# img = util.combine_img_prediction(data, label, prediction)
# util.save_image(img, "prediction.jpg")
```
